```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head(50)
```

```python
train = pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True) # Encode these columns with 0 and 1
```

```python
train.head()
train["Sex_male"] = train["Sex_male"].astype(int)
train["Embarked_Q"] = train["Embarked_Q"].astype(int)
train["Embarked_S"] = train["Embarked_S"].astype(int)
train.head()


train.drop("Name", axis =1)

train['Has_Cabin'] = train['Cabin'].notna().astype(int)
```

```python
tr = train.copy()
tr = tr.drop(["Cabin"], axis=1)
tr = tr.drop(["Name"], axis=1)
tr = tr.drop(["Ticket"], axis=1)
```

```python
# EDA
col = tr.columns
tr.info()       ## We see that there are missing values for age. Handle with data imputation.
missing_count = tr.isnull().sum()

import seaborn as sns 
import numpy as np

# Visualizations 

tr.head() 

""" 
The goal is to find correlations to survival rate. There could be specific
characteristics that predict suvival such as age, embarked, cabin, or class (i.e. wealthy or not). 
Packages used: seaborn, matplotlib

""" 

corr = tr.corr() 
sns.heatmap(corr, cmap = "inferno", annot=True, fmt=".1f", linecolor="white", linewidths=1)

""" 
The heatmap shows the relationships between survived and other columsns. We will
only keep those that are significant (i.e |r| >= 0.2) in predicting survival rate. This implies: 
Pclass, Has_Cabin, Sex_male, Fare, Embarked_S. 

"""
```

```python
# Model Selection 

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
X_train = tr(['Survived', 'PassengerId', 'Age','SibSp', 'Parch','Embarked_Q'], axis=1)
y_train = tr["Survived"]
X_test = test.drop(["PassengerId",'Age','SibSp', 'Parch','Embarked_Q'], axis=1) # Does not include survived
```

```python
# Needs numeric only
```

```python
# Imputation

imputer = SimpleImputer(missing_values=np.nan, strategy="median")
X_train_num = pd.DataFrame(imputer.fit_transform(X_train), columns = X_train.columns)
X_test_num = pd.DataFrame(imputer.transform(X_test), columns = X_test.columns)
```

```python
# Scaling
scaler = StandardScaler()

X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train_num), columns = X_train_num.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test_num), columns = X_test_num.columns)
```

```python
# Modeling

### Logistics Regression 

log = LogisticRegression() 

log.fit(X_train_scaled, y_train)
```

```python
y_pred = log.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
### Decision Tree Classifier 

from sklearn.tree import DecisionTreeClassifier
```

```python
DecisionTree = DecisionTreeClassifier(max_depth=5) 
```

```python
DecisionTree.fit(X_train_scaled, y_train)
```

```python
tree_predict = DecisionTree.predict(X_test_scaled)
```

```python
### KNN Classification 

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier() 
```

```python
knn.fit(X_train_scaled, y_train)
```

```python
knn_predict = knn.predict(X_test_scaled)
```

```python
### Support Vector Machine 

from sklearn.svm import SVC
```

```python
support = SVC(kernel="poly")

support.fit(X_train_scaled, y_train)
sp_pred = support.predict(X_test_scaled)
```

```python
## Employ Cross Validation 
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# CV 

log_scores = cross_val_score(log, X_train_scaled, y_train, cv=5) ## Noticable improvements 
tree_scores = cross_val_score(DecisionTree, X_train_scaled, y_train, cv=5)
knn_scores = cross_val_score(knn, X_train_scaled, y_train, cv=5)
svm_scores = cross_val_score(support, X_train_scaled, y_train, cv=5)
```

```python
print("Accuracy scores for each fold (Logistic):", log_scores)
print("Accuracy scores for each fold (Tree):", tree_scores)
print("Accuracy scores for each fold (KNN):", knn_scores)
print("Accuracy scores for each fold (SVM):", svm_scores)
```

```python
# Print MAX scores

print("Maximum score (Log):", max(log_scores))
print("Maximum score (Tree):", max(tree_scores))
print("Maximum score (KNN):", max(knn_scores))
print("Maximum score (SVM):", max(svm_scores))
```

```python
# We see that the KNN and SVM models both have the highest max scores from Cross Validation. This would be for the same default random state.
```

```python
# Feature importance (i.e what are the best features to use for the model)

from sklearn.feature_selection import RFE

selector = RFE(DecisionTree, n_features_to_select = 2)
selector.fit(X_train_scaled, y_train)
selected_features = X_train_scaled.columns[selector.support_]
print("Selected Features:", list(selected_features))
```
